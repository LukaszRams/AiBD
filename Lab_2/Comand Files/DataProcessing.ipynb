### Wczytanie danych

```python
data = pd.read_csv(r"../Original Data/earthquake_data.csv")  # załadowanie danych 
data.head(5)  # wyświetlenie pierwszych pięciu rekordów
```

Utworzenie nowego DataFrame z kolumnami:

- **Age** (stare: *Age*)

- **Gender** (stare: *What is your gender?*)

- **Answer** (stare: *Do you think the \"Big One\" will occur in your lifetime?*)

```python
# utworzenie nowego DataFrame z trzema kolumnami o nazwah Age, Gender, Answer na podstawie danych z oryginalnych danych
df = pd.DataFrame({"Age": data["Age"], 
                   "Gender":data["What is your gender?"], 
                   "Answer": data["Do you think the \"Big One\" will occur in your lifetime?"]})
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Usunięcie wierszy z niekompletnymi danymi (*NaN*)

```python
df = df.dropna()  # usunięcie wierszy z brakującymi danymi
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Utworzenie dwóch nowych kolumn:

- **Positive** - docelowo: liczba odpowiedzi "Yes" dla danej grupy wiekowej i płci

- **Negative** - docelowo: liczba odpowiedzi "No" dla danej grupy wiekowej i płci

```python
df["Positive"] = df["Answer"].mask(df["Answer"]=="No")  # utworzenie kolumny zawierającej "Yes" jeśli odpowiedź na pytanie była "Yes" lub NaN w przeciwnym przypadku 
df["Negative"] = df["Answer"].mask(df["Answer"]=="Yes")  # utworzenie kolumny zawierającej "No" jeśli odpowiedź na pytanie była "No" lub NaN w przeciwnym przypadku
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Pogrupowanie danych według **wieku** (kolumna: *Age*), a następnie **płci** (kolumna: *Gender*), a następnie obliczenie ilości danych w kolumnach *Answer*, *Positive Answer*, *Negative Answer*

```python
df = df.groupby(["Age", "Gender"]).count()  # pogrupowanie danych wg kolumn Age, Gender oraz obliczenie danych w pozostałych kolumnach
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Usunięcie kolumny **Answer**

```python
df.pop("Answer")
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Zresetowanie indeksów, tak aby były kolumnami

```python
df = df.reset_index(level=["Age", "Gender"])  # resetowanie indeksów
df.head(10)  # wyświetlenie pierwszych dziesięciu rekordów
```

Zapis DataFrame do pliku csv

```python
df.to_csv(r"..\Analysis Data\Earthquake_data_answer.csv", sep=",", index=False)
```
